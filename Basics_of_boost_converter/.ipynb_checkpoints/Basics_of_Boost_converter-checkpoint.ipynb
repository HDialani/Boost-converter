#  1 Read me

This is a simplified explanation of what a boost converter is for people without a bachelor of Electrical Engineering knowledge. What a Boost converter does is boost up the input voltage while boosting down the input current, since power in= power out. It does this by controlling the electronic switch(S).

# 2 How it works

When the switch is closed a large current $I$ flows through the inductor(L) and switch(S) causing the inductor to start storing magnetic field $\frac{1}{2}LI^{2}$. The moment that the switch opens the current flowing through the inductor will decrease since a load is now limiting the current. This change doesn't happen instantaneously since the inductor has some "current delay". As the current decreases the stored magnetic field will collapse, which causes the stored magnetic field energy to be released. Now that the inductor is releasing the energy it can be seen as an additional voltage source in series to the input voltage source. So during the period that the switch is open both the input voltage source and the collapsed voltage produced by the inductor go to the output and the collapsed voltage is larger than the input voltage. These voltages are then stored at the capacitor(C).  

When the switch is open the voltage at the output is equal to the voltage over the capacitor, when the switch is closed the capacitor slowly discharges since it has a "voltage delay". This causes the voltage at the output to slowly decrease until the switch is open again. The switch is opening and closing at a super-fast rate>100 kHz so the sudden switching in output voltage is not noticed and instead the average voltage will be noticed at the output.

Now for the final component the diode (D). The diode can be seen as a wire that allows current to flow in the direction of the arrow. When the switch is open the voltage over the cap is larger than the voltage of the input source, so the capacitor will try to discharge the current towards the input voltage. To stop this the diode makes sure that current can't flow back into the input source and instead goes to the output.

As for the output current, only when the switch is open current goes to the output, when the switch is closed it flows back into the battery. So the output current will be less than the input current.

##  3.1 Continuous conduction mode (CCM),Boundary condition mode (BCM) and Discontinuous conduction mode (DCM)

As mentioned before current  going through the inductor will be rising when the switch is closed and discharging when the switch is open. If the current rises and discharges above $0A$ it is in CCM. If it rises and discharges at $0 A$ it is in BCM. Current can't go below $0 A$ for the system so instead in the BCM it will first start to discharge till it reaches $0 A$ and stays $0 A$ until it starts rising again.

#  4 Simple math and simulations explaining how it is possible that $V_{out}>V_{in}$

First some basic circuit theory.

| Passive component | During charging (Load) |During discharging (Generator) | Equation|
|------|------|------|------|
|   Capacitor | Voltage rises $V \uparrow$<br> Current goes in to the capacitor $I+$ | Voltage falls $V \downarrow$<br> Current goes from to the capacitor $(I-)$ | $I=C\frac{d V}{d t}$|
|   Inductor  | Current rises $I \uparrow$<br> Voltage over the inductor is  $V+$| Current falls $I \downarrow$<br> Voltage over the inductor is  $V-$|$V=L\frac{d I}{d t}$|

In the previous model the input capacitor was neglected, that was because this is assuming an ideal voltage source. In practice a capacitor needs to be placed at the input. This is to make sure that a steady voltage source is given at the input. Look up decoupling capacitors if you are interested

<img src="Images/Boost_circuit_with_input_cap.png" width=500 >

During the stage when (S) is closed current in flowing from the input capacitor towards to inductor. In this case the capacitor is a generator and the inductor is the load. Initially $V_{out}=0$ .After a few cycles the circuit on the right is formed where the output capacitor acting as a generator is providing current to the load. So $V_{out}$ is constant.

<img src="Images/Boost_circuit_with_input_cap_Switch_closed.png" width=500 >

Next when the switch is open. The input capacitor is still a generator, but inductor is also now a generator. Both are providing current to the output capacitor and load. This can be given as $V_{in}+V_{L}=V_{out}+V_{D}$. The voltage over the input cap is $V_{in}$ and the voltage over the output cap is $V_{out}$. In practice $V_{D}<<V_{in},V_{L},V_{out}$ for power electronics level. So the equation can be simplified to $V_{in}+V_{L}=V_{out}$. Since the inductor is acting as a generator it has a $>0 V$, so this means $V_{out}>V_{in}$. 

<img src="Images/Boost_circuit_with_input_cap_Switch_open.png" width=500 >

So $V_{out}$ can be larger than $V_{in}$. Keep in mind voltage is boosted, but current is decreased that is why this model can exist. So without any losses $P_{out}=P_{in}$.

## 5.1 CCM

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

###  5.1.1 Inductor
General equation for inductor $I_L=\frac{1}{L}\int_{t_2}^{t_1} V_L \,dt$  

When S is closed $V_L=V_{in}$    

When S is open $V_L=V_{in}-V_{out}$   

```python
# Parameters
F_s = float(8.0e04)   # Hz
V_out = int(150)   # V
V_in = int(45)   # V
Load = int(400)    # Ohm
Delta_I_Ratio = float(0.4)

D = 1 - V_in / V_out
I_out = V_out / Load  # A
I_in = V_out * I_out / V_in  # A
I_ripple = Delta_I_Ratio * I_in  # A
Vd = 0.6  # V
```

```python
# Sampling
S = 1000  # Sampling resolution
N = 4  # Number of signals, needs to be an even number
Ts = 1 / F_s  # s
X = np.arange(0, N * Ts, N * Ts / S)
```

```python
V_l = []
for i in range(N):
    Intermediate1 = np.array(np.ones(len(np.arange(0, D * Ts, N * Ts / S))) * V_in)
    V_l.extend(Intermediate1)
    Intermediate2 = np.array(np.ones(len(np.arange(Ts / S, (1 - D) * Ts, N * Ts / S))) * (V_in - V_out))
    V_l.extend(Intermediate2)

    
I_l = np.add(I_d, I_sw)    
```

```python
plt.figure(3)
plt.rcParams['figure.figsize'] = [30, 20]
plt.rcParams.update({'font.size': 22})
plt.subplot(211)
plt.plot(1e6 * X, V_l)
plt.xlim([0, N * 1e6 * Ts])
plt.ylim([V_in - V_out - 5, V_in + 5])
# plt.xlabel('Time [$\mu s$]')
plt.ylabel('$V_L$ [$V$]')

plt.subplot(212)
plt.plot(1e6 * X, I_l)
plt.xlim([0, N * 1e6 * Ts])
plt.ylim([-0.1, (I_in + I_ripple) + 0.1])
plt.xlabel('Time [$\mu s$]')
plt.ylabel('$I_{L}$ [$A$]')
plt.suptitle('Inductor')
```

### 5.1.2 Transistor
When S is closed $V_{S}=0$  and $I_{S}=I_L$ 

When S is open $V_{S}=V_{out}$  and $I_{S}=0$


```python
V_sw = []
for i in range(N):
        Intermediate1 = np.array(np.ones(len(np.arange(0, D * Ts, N * Ts / S))) * 0)
        V_sw.extend(Intermediate1)
        Intermediate2 = np.array(np.ones(len(np.arange(Ts / S, (1 - D) * Ts, N * Ts / S))) * (V_out + Vd))
        V_sw.extend(Intermediate2)

I_sw = []
for i in range(N):
        I_intermediate1 = np.array(I_in - I_ripple + (np.arange(0, D * Ts, N * Ts / S)) * 2 * I_ripple / (D * Ts))
        I_sw.extend(I_intermediate1)
        I_intermediate2 = np.array(np.ones(len(np.arange(Ts / S, (1 - D) * Ts, N * Ts / S))) * 0)
        I_sw.extend(I_intermediate2)
```

```python
plt.figure(1)
plt.rcParams['figure.figsize'] = [30, 20]
plt.rcParams.update({'font.size': 22})
plt.subplot(211)
plt.plot(1e6 * X, V_sw)
plt.xlim([0, N * 1e6 * Ts])
plt.ylim([-5, V_out + 5])
# plt.xlabel('Time [$\mu s$]')
plt.ylabel('$V_{sw}$ [$V$]')

plt.subplot(212)
plt.plot(1e6 * X, I_sw)
plt.xlim([0, N * 1e6 * Ts])
plt.ylim([-0.1, (I_in + I_ripple) + 0.1])
plt.xlabel('Time [$\mu s$]')
plt.ylabel('$I_{sw}$ [$A$]')
plt.suptitle('Transistor')
```

### 5.1.3 Diode
When S is closed $V_{D}=V_{out}$ and $I_{D}=0$ 

When S is open $V_{D}=V_{forward-drop}$ and $I_{D}=I_L$ 

```python
V_d = []
for i in range(N):
    Intermediate1 = np.array(np.ones(len(np.arange(0, D * Ts, N * Ts / S))) * -V_out)
    V_d.extend(Intermediate1)
    Intermediate2 = np.array(np.ones(len(np.arange(Ts / S, (1 - D) * Ts, N * Ts / S))) * (Vd))
    V_d.extend(Intermediate2)

I_d = []
for i in range(N):
    I_intermediate1 = np.array(np.arange(0, D * Ts, N * Ts / S) * 0)
    I_d.extend(I_intermediate1)
    I_intermediate2 = np.array(I_in + I_ripple - (np.arange(Ts / S, (1 - D) * Ts, N * Ts / S)) * 2 * I_ripple / ((1 - D) * Ts))
    I_d.extend(I_intermediate2)
```

```python
plt.figure(2)
plt.rcParams['figure.figsize'] = [30, 20]
plt.rcParams.update({'font.size': 22})
plt.subplot(211)
plt.plot(1e6 * X, V_d)
plt.xlim([0, N * 1e6 * Ts])
plt.ylim([-V_out + -5, 5])
# plt.xlabel('Time [$\mu s$]')
plt.ylabel('$V_{d}$ [$V$]')

plt.subplot(212)
plt.plot(1e6 * X, I_d)
plt.xlim([0, N * 1e6 * Ts])
plt.ylim([-0.1, (I_in + I_ripple) + 0.1])
plt.xlabel('Time [$\mu s$]')
plt.ylabel('$I_{d}$ [$A$]')
plt.suptitle('Diode')
```

### 5.1.4 Capacitor
When S is closed $V_C=V_{out}$ and $I_C=-I_{out}$ 

When S is open $V_C=V_{out}$ and $I_C=I_{D}-I_{out}$ 

```python
V_c = []
for i in range(N):
    Intermediate1 = np.array(np.ones(len(np.arange(0, D * Ts, N * Ts / S))) * V_out)
    V_c.extend(Intermediate1)
    Intermediate2 = np.array(np.ones(len(np.arange(Ts / S, (1 - D) * Ts, N * Ts / S))) * V_out)
    V_c.extend(Intermediate2)

I_c = []
for i in range(N):
    I_intermediate1 = np.array(-I_out - np.arange(0, D * Ts, N * Ts / S))
    I_c.extend(I_intermediate1)
    I_intermediate2 = np.array(-I_out + I_in + I_ripple - (np.arange(Ts / S, (1 - D) * Ts, N * Ts / S)) * 2 * I_ripple / ((1 - D) * Ts))
    I_c.extend(I_intermediate2)
```

```python
plt.figure(4)
plt.rcParams['figure.figsize'] = [30, 20]
plt.rcParams.update({'font.size': 22})
plt.subplot(211)
plt.plot(1e6 * X, V_c)
plt.xlim([0, N * 1e6 * Ts])
plt.ylim([V_out - 5, V_out + 5])
# plt.xlabel('Time [$\mu s$]')
plt.ylabel('$V_c$ [$V$]')

plt.subplot(212)
plt.plot(1e6 * X, I_c)
plt.xlim([0, N * 1e6 * Ts])
plt.ylim([-I_out - 0.1, (I_in + I_ripple - I_out) + 0.1])
plt.xlabel('Time [$\mu s$]')
plt.ylabel('$I_{c}$ [$A$]')
plt.suptitle('Capacitor')
plt.show()
```

## 5.2 DCM

The waveforms are based on [this article](https://www.researchgate.net/publication/328871868_DC-Transformer_Modelling_Analysis_and_Comparison_of_the_Experimental_Investigation_of_a_Non-Inverting_and_Non-Isolated_Nx_Multilevel_Boost_Converter_Nx_MBC_for_Low_to_High_DC_Voltage_Applications)
and 
[this video](https://www.youtube.com/watch?v=Uu_sJmlUFVE)

```python
# Parameters
F_s = float(8.0e04)   # Hz
V_out = int(150)   # V
V_in = int(45)   # V
Load = int(400)    # Ohm


L=0.1/1000
D=math.sqrt(((2*V_out**2/Load*L*F_s)*(V_out-V_in)/(V_out*V_in**2)))
Delta1=(2*(V_out/Load)*L*F_s)/(V_in*D)
Delta2=1-D-Delta1
```

```python
S=1000 # Sampling resoltion
N=2 #Number of signals
Ts=1/F_s #s
X= np.arange(0, N*Ts,N*Ts/S)
```

### 5.2.1 Inductor
General equation for inductor $I_L=\frac{1}{L}\int_{t_2}^{t_1} V_L \,dt$  

When S is closed $V_L=V_{in}$   

When S is open during $\Delta_1$ $V_L=V_{in}-V_{out}$  

When S is open during Zero $V_{L}=V_{0}$

```python
V_L=[]
V_intermediate1=[]
V_intermediate2=[]
V_intermediate3=[]

for i in range(N):
    V_intermediate1=np.array(np.ones(len(np.arange(0,D*Ts,N*Ts/S)))* V_in)
    V_L.extend(V_intermediate1)
    V_intermediate2=np.array(np.ones(len(np.arange(Ts/S,(Delta1)*Ts,N*Ts/S)))*(V_in-V_out))
    V_L.extend(V_intermediate2)
    V_intermediate3=np.array(np.ones(len(np.arange(Ts/S,(1-D-Delta1)*Ts,N*Ts/S)))*0)
    V_L.extend(V_intermediate3)    
```

```python
del V_L[-N:] # Ugly fix to make this a 1002 array instead of 1000 array
```

```python
I_l = np.add(I_d, I_sw) 
```

```python
plt.figure(3)
plt.rcParams['figure.figsize'] = [30, 20]
plt.rcParams.update({'font.size': 22})
plt.subplot(211)
plt.plot(1e6 * X, V_L)
plt.xlim([0, N * 1e6 * Ts])
plt.ylim([V_in - V_out - 5, V_in + 5])
# plt.xlabel('Time [$\mu s$]')
plt.ylabel('$V_L$ [$V$]')

plt.subplot(212)
plt.plot(1e6 * X, I_l)
plt.xlim([0, N * 1e6 * Ts])
plt.ylim([-0.1, (2*I_ripple) + 0.1])
plt.xlabel('Time [$\mu s$]')
plt.ylabel('$I_{L}$ [$A$]')
plt.suptitle('Inductor')
```

When S is open during $\Delta_1$ $V_{S}=V_{out}$  and $I_{S}=0$

When S is open during Zero $V_{S}=V_{in}$  and $I_{S}=0$

```python
V_sw=[]

for i in range(N):
    V_intermediate1=np.array(np.ones(len(np.arange(0,D*Ts,N*Ts/S)))* 0)
    V_sw.extend(V_intermediate1)
    V_intermediate2=np.array(np.ones(len(np.arange(Ts/S,(Delta1)*Ts,N*Ts/S)))*(V_out))
    V_sw.extend(V_intermediate2)
    V_intermediate3=np.array(np.ones(len(np.arange(Ts/S,(1-D-Delta1)*Ts,N*Ts/S)))*(V_in))
    V_sw.extend(V_intermediate3)   
```

```python
del V_sw[-N:]# Ugly fix to make this a 1002 array instead of 1000 array
```

### 5.2.3 Diode
When S is closed $V_{D}=V_{out}$ and $I_{D}=0$ 

When S is open during $\Delta_1$ $V_{D}=V_{forward-drop}$ and $I_{D}=I_L$

When S is open during Zero $V_{S}=V_{-in}$  and $I_{S}=0$

### 5.2.4 Capacitor
When S is closed $V_C=V_{out}$ and $I_C=-I_{out}$ 

When S is open during $\Delta_1$ $V_C=V_{out}$ and $I_C=I_{D}-I_{out}$ 

When S is open  during Zero $V_C=V_{out}$ and $I_C=-I_{out}$ 

```python
I_out=1
```

```python
V_c = []
for i in range(N):
    V_intermediate1=np.array(np.ones(len(np.arange(0,D*Ts,N*Ts/S)))*V_out)
    V_c.extend(V_intermediate1)
    V_intermediate2=np.array(np.ones(len(np.arange(Ts/S,(Delta1)*Ts,N*Ts/S)))*(V_out))
    V_c.extend(V_intermediate2)
    V_intermediate3=np.array(np.ones(len(np.arange(Ts/S,(1-D-Delta1)*Ts,N*Ts/S)))*(V_out))
    V_c.extend(V_intermediate3) 

I_c = []
for i in range(N):
    I_intermediate1=np.array( np.arange(0,D*Ts,N*Ts/S)* 0-I_out)
    I_c.extend(I_intermediate1)
    I_intermediate2=np.array( 2*I_ripple-(np.arange(Ts/S,(Delta1)*Ts,N*Ts/S))* 2 * I_ripple /((1-D-Delta1+0.04)* Ts)-I_out )
    I_c.extend(I_intermediate2)
    I_intermediate3=np.array( (np.arange(Ts/S,(1-D-Delta1)*Ts,N*Ts/S))* 0-I_out)
    I_c.extend(I_intermediate3)
```

# 7 When can it enter DCM or CCM/BCM?

5 parameters can change the Boost operation mode: inductor value, load resistance, output voltage, switching frequency, and duty cycle. This part shows how each of them can affect this.

## 7.1 Load resistor

Assuming everything else remains the same. Increasing the load resistance can push the Boost converter to enter discontinuous mode. The relationship is that as load resistance increases the output current will decrease, which means the average current of the inductor will decrease. The load resistance does not affect the ripple current, however, so if the average input current is less than the ripple current it will go to zero and stay zero till the inductor starts to charge again. The boundary condition is $I_{in}=\frac{\Delta I_L}{2}$. If $I_{in}>\frac{\Delta I_L}{2}=>\frac{V_{out}}{R}\cdot \frac{1}{1-d}>\frac{\Delta I_L}{2}$ then it is CCM, the other way around is DCM. So increasing R and keeping everything else the same pushes the boost converter more toward DCM

<p float="left">
  <img src="Images/CCM_DCM_R_Load_CCM_Mode.jpg" width=400 > 
  <img src="Images/CCM_DCM_R_Load_DCM_Mode.jpg" width=400 > 
</p>

 ## 7.2 Output voltage
 
 Since $V=I*R$ if R remains the same but V decreases, I will also decrease till the converter enters DCM

## 7.3 Inductance 

For inductance, you are manipulating $\Delta I_L$, while keeping $I_{in}$ the same. 
This is due to the equation $\Delta I_L=\frac{V_{in}}{L}DT_{S}$
Changing $L$ doesn't effect $I_{in}$. 
So this time increasing the inductance can result in the boost converter going from CCM to DCM mode and decreasing it can mode the converter from DCM to CCM mode.

$L_{crit}=\frac{ V_{in}^2DR}{2f_{sw}V_{out}^2}$ this [video](https://www.youtube.com/watch?v=YIg7J3YjnbM&list=PLmK1EnKxphinxBub5hL0ZoJXWoqjkGE19&index=33&t=3) shows the entire derivation if you are interested


If $L>L_{crit}$ you are in CCM mode. <br />
If $L< L_{crit}$ you are in DCM mode..

## 7.4 Switching frequency
From above $L_{crit}=\frac{ V_{in}^2DR}{2f_{sw}V_{out}^2}$. So decreasing $f_{sw}$ will cause $L_{crit}$ to increase. 

## 7.5 Duty cycle
If D increases, $L_{crit}$ increases

# 8 Power supply design 

The whole point of a power supply is that when you ask for a specific voltage you want to get that voltage at the output. For a boost converter you expect the DC voltage to be equal to the request voltage. Obviously you can't get the exact value due, but you still expect a relatively small error. The specific value depends on the quality and application purpose of the power supply, but I assume max $5\%$. In practice you would never use a boost converter as your own lab power supply since you need a large range of voltages so a buck-boost would be more useful. But since isolation from high voltage is also extremely important a flyback topology is chosen as the most commonly used lab bench power supply. The [ESP lab](https://www.tudelft.nl/ewi/onderzoek/faciliteiten/esp-lab) that gave me the equipment to do this " educational experiment" only had the boost converter so I made this entire project based on that.   

When using a power supply the user want to just supply voltage to a specific load and not care about mode of operation, $f_{sw}$, internal inductor used etc. The user will not touch these values. What variables need to be modified in order to get the user desired voltage? The user only has control over the load and in many cases don't know the resistance of the load and doesn't bother in finding it out when using a power supply. They don't need to care about the load's resistance the power supply just needs to accept it. So what parameters should be modified if a new voltage is asked? Well if it is in CCM obviously the duty cycle. But what if the load's resistance is so high that the boost converter enters DCM? The user doesn't care about the operation mode of the boost converter. The easy way is to design the boost converter so that it that it will almost always act in CCM, but I hate mental health so I want to make mine be able to operate in both CCM and DCM. 

So in DCM what would you modify? Here is the equation for $V_{out}$ in DCM  $V_{out}=\frac{V_{in}}{2}+\frac{V_{in}}{2}\sqrt{1+\frac{2 D^2 R}{f_{sw} L}} $ [Source](https://www.youtube.com/watch?v=6RuPplUim4E&t=3s). The inductor (L)? Since you need to either change the number of winding or modify the air gap of a core in order to change the chosen inductors inductance it is not likely the ever change when the power supply is being used. The switching frequency? The switching frequency could be modified but in many cases can also be fixed. If it is too low $<20kHz$ it becomes quite audible. For younger children the minimum frequency can be higher. If it is too high switching losses can be massive and thus you lose efficiency and or need to over design the cooling needed. Just as mentioned above the load (R) can't be modified into  since this is a user dependent parameter. The input voltage $V_{in}$? In some cases yes but in majority of the scenario this is also fixed by another designer. So that only leaves duty cycle (D). You might be wondering since I already mention  that the R is usually not known during operation so how can the correct $V_{out}$ be calculated without this parameter? Well in practice a feedback system is added that checks the output voltage and then modifies the micro-controller's duty cycle till the desired output voltage is achieved. In my design this is also present. In the CCM mode the feedback system is also present since calculated $V_{out}$ and practice $V_{out}$ can still be different even when taking losses into account. 

So the now the questions becomes how to choose the correct duty cycle during DCM mode? And how would the duty-cycle's value look like for a given a specific boost specification. Well welcome to the end this last section will go over these specifically.


## CCM mode and Duty cycle

In an ideal world you would get the following relationship between $V_{out}$ and $D$. As $D$ approaches 1 $V_{out}$ approaches $\infty$

```python
V_in = int(45) 
D_length = 1000
D = np.linspace(0, 0.99, D_length)
V_out=V_in/(1-D)
```

```python
plt.plot(D, V_out )
plt.xlabel("Duty cycle (D)")
plt.ylabel("Output voltage in $V$")
plt.title('Ideal Boost converter')
plt.xlim([0, 1])
plt.ylim([0, 1000])
```

But in the real world there are parasitics and the circuit will look more like the following

<img src="Images/Boost_converter_with_parasitics.png" width=500 > 

[Source](https://www.researchgate.net/publication/273392829_Sliding-Mode-Control-Based_Boost_Converter_for_High-Voltage-Low-Power_Applications)

Now I could try to derive the $V_{out}$(D) for this model, but thankfully the [Source](https://www.researchgate.net/publication/273392829_Sliding-Mode-Control-Based_Boost_Converter_for_High-Voltage-Low-Power_Applications) did that already for me :) The equation is as followed.

$\frac{V_o}{V_S}=\frac{R_o(1-D)}{R_o(1-D)^2+\frac{D(1-D) \cdot R_o \cdot r_C}{R_o+r_C}+r_{\mathrm{ON}}+r_L}$

By setting $r_C,r_{\mathrm{ON}}$ and $r_L$ to $0$ you get the ideal model. 

```python
V_in = int(45) 
D_length = 1000
D = np.linspace(0, 0.99, D_length)

R_load=400

Rc=0;
Ron=0;
Rl=0;
V_out_ideal=(V_in*R_load*(1-D))/(D*(1-D)*R_load*Rc/(R_load+Rc)+Ron+Rl+R_load*(1-D)**2)

Rc=30/1000;
Ron=0.01;
Rl=300/1000;
V_out_parasitics=(V_in*R_load*(1-D))/(D*(1-D)*R_load*Rc/(R_load+Rc)+Ron+Rl+R_load*(1-D)**2)
```

```python
plt.plot(D, V_out_ideal, color='r', label='Ideal' )
plt.plot(D, V_out_parasitics, color='b', label='Parasitics' )
plt.xlabel("Duty cycle (D)")
plt.ylabel("Output voltage in $V$")
plt.title('Boost converter in CCM')
plt.xlim([0, 1])
plt.ylim([0, 1000])
plt.legend()
```

## DCM mode and Duty cycle

### Keeping everything the same and modifying the duty cycle only

Boundry

$I_{L,boundry}=\frac{1}{2}I_{peak}$

$I_{L,boundry}=\frac{V{in}}{2 L}t_{on}$

$I_{L,boundry}=\frac{V{out}}{2 L}(1-D)t_{on}$

$I_{L,boundry}=\frac{V{out}}{2 L f_{sw}}D\cdot(1-D)$

$I_{out,boundry}=\frac{V{out}}{2 L f_{sw}}D\cdot(1-D)^2$

$\frac{I_{out,boundry}}{I_{out}}=\frac{R}{2 L f_{sw}}D\cdot(1-D)^2$, at boundry $\frac{I_{out,boundry}}{I_{out}}=1$

So now I can rewrite the boundry condition as

$\frac{2 L f_{sw}}{R}=D\cdot(1-D)^2$

$\frac{2 L f_{sw}}{R}>D\cdot(1-D)^2$ for CCM 

$\frac{2 L f_{sw}}{R}<D\cdot(1-D)^2$ for DCM 

 Below gives a nice visual reprentation as to what happens when you only modify the duty cycle, but keep everything else the same.
 <img src="Images/CCM_DCM_K_crit_graph.jpg" width=300 > 
source left:https://ninova.itu.edu.tr/en/courses/institute-of-science-and-technology/132/elk-506e/ekkaynaklar?g45743 <br />

Here $K_{crit}=D*(1-D)^2$ and $K=\frac{2Lf_{sw}}{R}$

```python
def operation_mode(L,F_s,Load):
   D_length = 100
   D_0_to_1 = np.linspace(0, 1, D_length)
   K_crit = D_0_to_1 * (1 - D_0_to_1) ** 2

   K = 2 * L * F_s / Load
   K = K * np.ones(D_length)
   fig = plt.figure()
   plt.plot(D_0_to_1, K_crit, color='r', label='K_crit', )
   plt.plot(D_0_to_1, K, color='g', label='K')
   plt.xlabel("Duty cycle")
   plt.legend()
   plt.title('If K>K_crit you are in CCM and if K<K_crit you are in DCM ')
   return fig
```

```python
L=0.75/1000
F_s=5*10**3
Load=400
K=operation_mode(L,F_s,Load)
plt.show()
```

$I_{\mathrm{out}_{-} \mathrm{Boundry}}=\frac{V{out}}{2 L f_{sw}}D\cdot(1-D)^2$

```python
F_s = float(8.0e04) # Hz
V_out = int(150)    # V
L=0.75/1000         # H
```

```python
D_length = 1000
D = np.linspace(0, 1, D_length)
I_Out_Boundry=V_out/(2*F_s*L)*D*(1 - D)**2
```

```python
plt.plot(D, I_Out_Boundry, color='r', label='I_out boundry current' )
plt.xlabel("Duty cycle")
plt.ylabel("I in $A$")
plt.title('$V_{out}$ is constant')
plt.legend()
```

Which duty cycle gives the highest possible $I_{out,boundry}$?
